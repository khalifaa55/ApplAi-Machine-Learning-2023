### Task 3: Checking missing values and Outliers & Creating visual methods to analyze the data.

```python
df.isnull().sum()
```

```python
total_cells = df.size
null_cells = df.isnull().sum()
print((null_cells/total_cells)*100)
```

```python
df['bmi'].fillna(df['bmi'].mean(), inplace=True)
df.drop("id", axis=1, inplace=True)
df['gender'].unique()
```

```python
df[df['age']<1]
df= df[df['age']>1]
```

```python
df['gender'].value_counts().get('Other', 0)
df['gender']=df['gender'].replace('Female',0)
df['gender']=df['gender'].replace('Male',1)
df = df[df['gender'] != 'Other']
```

```python
df['smoking_status'].value_counts().get('Unknown',0)
#df= df[df['smoking_status']!= 'Unknown']
df['smoking_status'].unique()
df['smoking_status']=df['smoking_status'].replace('never smoked',0)
df['smoking_status']=df['smoking_status'].replace('smokes',1)
df['smoking_status']=df['smoking_status'].replace('formerly smoked',2)
df['smoking_status'].replace('Unknown', df['smoking_status'].mode()[0], inplace=True)
```

```python
df['ever_married'].unique()
df['ever_married']=df['ever_married'].replace('Yes',1)
df['ever_married']=df['ever_married'].replace('No',0)
```

```python
df['Residence_type'].unique()
df['Residence_type']=df['Residence_type'].replace('Urban',0)
df['Residence_type']=df['Residence_type'].replace('Rural',1)
```

### Task 4: creat a model that fits the data

```python
X = df.drop('stroke',axis=1)
y = df['stroke']
```

```python
#Logistic Regression Model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 41)
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
y_pred = classifier.predict(X_test)
cr = classification_report(y_test, y_pred, zero_division=1) 
print(cr)
```

```python
#KNN Model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
model=KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
y_preds=model.predict(X_test)

print(f"KNN testing score = {accuracy_score(y_test,y_preds)}")
```

```python
#KNN Model with hyperparameter tuning
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
# Define the parameter grid for Random Search
param_grid = {
    'n_neighbors': np.arange(1, 21),    # Range of values for the number of neighbors
    'weights': ['uniform', 'distance'],  # Different weighting schemes
    'p': [1, 2],                          # Power parameter for Minkowski distance (1=Manhattan, 2=Euclidean)
}

# Create a KNN classifier
knn = KNeighborsClassifier()

# Create a RandomizedSearchCV object
random_search = RandomizedSearchCV(knn, param_distributions=param_grid, n_iter=10, cv=5, random_state=42)

# Perform the random search on the training data
random_search.fit(X_train, y_train)

# Print the best hyperparameters found
print("Best Hyperparameters:", random_search.best_params_)

# Evaluate the model on the test data using the best hyperparameters
stroke_model = random_search.best_estimator_
accuracy = stroke_model.score(X_test, y_test)
print("Test Accuracy:", accuracy)
cm = confusion_matrix(y_test, y_pred)
print("confusion matrix:\n ", cm)
```
